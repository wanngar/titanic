# Обучение нейронной сети

Вынесем целевую переменную отдельно

```python
X = DF.drop(columns=['Survived'])
y = DF['Survived']
```

Разделим на обучающую и тестовую выборки

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y.shape[0]
```

## Первая модель

```python
model_v1 = Sequential()
model_v1.add(Dense(32, activation='relu', input_shape=(X.shape[1],)))
model_v1.add(Dense(16, activation='relu'))
model_v1.add(Dense(1, activation='sigmoid'))
```

```python
model_v1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history  = model_v1.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model v1')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# Evaluate the model
loss, accuracy = model_v1.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

## Вторая модель

```python
model_v2 = Sequential()
model_v2.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],),kernel_regularizer='l2'))
model_v2.add(Dropout(0.3))
model_v2.add(Dense(32, activation='relu',kernel_regularizer='l2'))
model_v2.add(Dropout(0.3))
model_v2.add(Dense(16, activation='relu',kernel_regularizer='l2'))
model_v2.add(Dense(1, activation='sigmoid'))
```

```python
model_v2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model v2')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# Evaluate the model
loss, accuracy = model_v2.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
model_v3 = Sequential()
model_v3.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],),kernel_regularizer='l2'))
model_v3.add(Dropout(0.3))
model_v3.add(Dense(128, activation='relu',kernel_regularizer='l2'))
model_v3.add(Dropout(0.3))
model_v3.add(Dense(64, activation='relu',kernel_regularizer='l2'))
model_v3.add(Dropout(0.3))
model_v3.add(Dense(32, activation='relu',kernel_regularizer='l2'))
model_v3.add(Dropout(0.3))
model_v3.add(Dense(16, activation='relu',kernel_regularizer='l2'))
model_v3.add(Dense(1, activation='sigmoid'))
```

## Кросс валидация

```python
def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
acc_list = []
c=1
for train_i, test_i in kf.split(X):

    X_tr, X_ts = X.iloc[train_i], X.iloc[test_i]
    y_tr, y_ts = y.iloc[train_i], y.iloc[test_i]

    model = model_v2
    history = model.fit(X_tr, y_tr, batch_size = 32, epochs = 50,
        validation_split=0.2,  shuffle = True, verbose = 0)

    loss, accuracy = model.evaluate(X_ts, y_ts, verbose = 0)
    print(f'fold №{c}')
    print(f'Test Accuracy: {accuracy:.4f}')

    #вывод метрики accuracy и loss
    deeplab_accuracy = history.history['accuracy']
    deeplab_val_accuracy = history.history['val_accuracy']
    deeplab_loss = history.history['loss']
    deeplab_val_loss = history.history['val_loss']


    #получаем предсказания модели
    y_pred = model.predict(X_ts)

    y_pred_binary = np.where(y_pred > 0.5, 1, 0)
    plot_confusion_matrix(y_ts, y_pred_binary)


    #выводим classification report
    print(classification_report(y_ts, y_pred_binary))

    #для вывода средней точности
    rounded_accuracy = round(accuracy, 2)
    acc_list.append(rounded_accuracy)
    c+=1

#считаем среднюю точность по всем фолдам
mean_accuracy = np.mean(acc_list)
print(f'Test_mean_accuracy: {mean_accuracy:.4f}')
```

Модель номер 2 на кросс валидации из 5 фолдов показала максимальную точность 87%

В целом, модель показывает хорошую общую производительность, но может потребовать улучшения в отношении чувствительности к положительным классам для достижения оптимального баланса между точностью и обнаруживаемостью всех классов.
